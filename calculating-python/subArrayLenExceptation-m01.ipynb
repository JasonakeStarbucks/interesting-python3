Given a string consists of '0' and '1', find the expectation of the longest length of 'all-1' or 'all-0' substring.

```python
import numpy as np

def generateString(m: int) -> int:
    """
    :param m: length of string
    :return: the longest length of 'all-0/1' substring of this string
    """
    maxLength = 0
    oppostite = 0
    tag = 0 # tag = 0 means 0 is the element of current longest string
    for i in range(m):
        if np.random.rand(1) < 0.5: # generate 0
            if tag == 0:
                maxLength = maxLength + 1
            else:
                oppostite = oppostite + 1
        else:
            if tag == 1:
                maxLength = maxLength + 1
            else:
                oppostite = oppostite + 1
        if oppostite == maxLength:
            tag = 1 - tag
            oppostite = 0
    return maxLength
```

```python
x = np.arange(1, 101, 1, dtype='int')
y = np.zeros((1, 150))
X, Y = np.meshgrid(x, y)

for (i, j), value in np.ndenumerate(X):
    Y[i, j] = generateString(value)

X = X.ravel()
Y_mean = np.mean(Y, axis=0)
Y = Y.ravel()

import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X, Y, s=5)
plt.plot(x, Y_mean, c='r')
plt.show()

(x[-1], Y_mean[-1])
```
